# How should binary inputs and outputs be encoded in a neural network?

## Exploring options with a toy problem in binary logic using Keras.

Like many people who play with artificial neural networks (ANNs), I did some googling about best practices for input and output encoding in neural networks, activation functions, weight initialization, etc. Most sources were in fairly good agreement about how, when, and why to normalize/standardize numeric inputs and targets and the importance of matching inputs to the domain of the activation functions in the first hidden layer and matching the range of the output activation functions to the expected ANN output. For example, [here](https://visualstudiomagazine.com/articles/2013/07/01/neural-network-data-normalization-and-encoding.aspx)'s a link to an article recommending the following rules of thumb:

> 1. Binary X data: use -1.0 or +1.0 encoding
2. Binary Y data: use 0.0 or 1.0 encoding
3. Categorical X data: use effects encoding such as [0.0, 0.0, 1.0] or [-1.0, -1.0, -1.0]
4. Categorical Y data: use dummy encoding such as [0.0, 0.0, 1.0] or [1.0, 0.0, 0.0]
5. Numeric X data: use (x - mean) / stddev normalization
6. Numeric Y data: do not normalize 

None of the above rules are written in stone, and the best results will depend very much on the specifics of the problem at hand. In this notebook, I will explore rules 1 and 2 using Keras to solve a simple binary logic problem.

## Numeric input encoding
Most sources I came across agree with rule 5: Numeric input variables should be Z- or min-max standardized. We want the input data to be zero-centered with a standard deviation of about 1, or at least get somewhere close to that. This is not because ANNs are incapable of solving problems with large-magnitude inputs (they certainly are), but because training is more efficient this way, especially if the inputs have large differences in their variances (inputs with larger variances will receive more focus in the weight updates at the expense of inputs with smaller variances) or magnitudes (inputs with larger magnitudes may saturate the input activation function into regions of vanishing derivatives, which inhibits training. If the initial random weights cause this saturation to occur in the wrong direction, convergence can be severely impaired. See LeCun's *Efficient Backprop* for more detail.) The training algorithm may eventually overcome both of these problems, but unless great care is taken to optimize the initial weights, convergence can be slow. Standardizing the inputs removes these problems and allows us to initialize the weights with small random values. When both numeric and binary input variables are used, it becomes even more important to standardize the numeric inputs to similar magnitudes and variances.

As for zero-centering, LeCun says:
> When all of the components of an input vector are positive, all fo the updates of weights that feed into a node will be the same sign. As a result, these weights can only all decrease or all increase together for a given input pattern. Thus, if a weight vector must change direction it can only do so by zigzagging which is inefficent and thus very slow.

Thus, standardizing numeric inputs with Z- or zero-centered min-max standardization is easy to do and improves training efficiency. 

## Binary input encoding
As with numeric input encoding, backprop is more efficient when the input variables are centered around zero. As such, LeCun recommends encoding true and false as -1 and 1 (ie. 1 of C-1 encoding). The more intuitive 0 and 1 encoding results in an positively skewed input variable. Furthermore, if tanh or sigmoid activation is used, vanishing inputs push the activation function toward its linear region, while non-vanishing inputs push the activation function toward nonlinearity, thus the activation function treats 0 and 1 asymmetrically, but treats -1 and 1 symmetrically. (Likewise, classification inputs can be encoded with 1 of C-1 encoding (aka effect encoding) for the same reason.)

To explore this, lets look at a binary logic problem using the keras library in python to create a simple ANN. I am using a 5-year-old Dell laptop with eight 2.2 GHz cores, and using the Theanos backend for Keras in python 2. I am using my GPU for training the neural net. Here is the code:

```python
# import libraries
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import matplotlib.pyplot as plt
from timeit import default_timer as timer

def binary_ANN(in_true, in_false, out_true, out_false):

    # generate training data 
    #(XOR, or any other binary logic truth table)
    dataModel = [[(in_false, in_false), (out_false)],
                 [(in_false, in_true), (out_true)],
                 [(in_true, in_false), (out_true)],
                 [(in_true, in_true), (out_false)]]


    train_size = 1000     # number of training inputs
    np.random.seed(1)

    # add rows randomly from the truth table
    def make_data(size):
        inp = []
        targ = []
        for ind in range(size):
            inpt, target = dataModel[np.random.randint(4)]
            inp.append(inpt)
            targ.append(target)
        return np.array(inp), np.array(targ).reshape(-1, 1)

    train_in, train_target = make_data(train_size)

    # 1. Define Network:
    num_hidden_nodes = train_in.shape[1] + 1
    model = Sequential()
    # 3 nodes in the hidden layer with tanh activation function
    model.add(Dense(num_hidden_nodes, input_dim=train_in.shape[1], activation='tanh', init='uniform'))
    # 1 output node with logistic sigmoid activation function
    model.add(Dense(1, activation='sigmoid', init='uniform'))

    # 2. Compile Network
    model.compile(loss='mse', optimizer='adadelta', metrics=['accuracy'])
    t = timer()

    # 3. Fit Network
    history = model.fit(train_in,
                        train_target,
                        nb_epoch=1000,
                        validation_split=0.2,
                        batch_size=32,
                        verbose=0)

    # How long did training take?
    print 'Time elapsed:', timer() - t

    # Visualize convergence
    %matplotlib inline
    train_err = history.history['loss']
    train_acc = history.history['acc']
    val_err = history.history['val_loss']
    val_acc = history.history['val_acc']

    plt.semilogy(range(len(train_err)), train_err, label='train')
    plt.semilogy(range(len(val_err)), val_err, label='val')
    plt.title('error')
    plt.legend(loc=3)
    plt.xlabel('epoch')
    plt.ylabel('Mean squared error')

    # Make predictions
    demo_in = np.array([element[0] for element in dataModel])
    demo_target = np.array([element[1] for element in dataModel])
    demo_out = model.predict(demo_in)
    misfit = demo_out-demo_target.reshape(-1,1)

    print 'Prediction:'
    for i, val in enumerate(demo_in):
        print demo_in[i], demo_out[i]

    print "\nMisfit:\n", misfit

    print "\nFinal mean squared error:", train_err[-1]
    
    return train_err, val_err
```

Let's look at the convergence curve and the prediction accuracy when we use standard binary encoding (True = 1, False = 0).

```python
train_err, val_err = binary_ANN(1,0,1,0)
```

Result: With traditional binary inputs and outputs, we get convergence to an MSE of ~2 x 10<sup>-4</sup> in 1000 iterations, with about two decimal places of precision in the predictions.

Now let's change to True = 1, False = -1 for the inputs only:

```python
train_err, val_err = binary_ANN(1,-1,1,0)
```

Result: With zero-centered binary inputs and outputs, we get convergence to an MSE of ~7 x 10<sup>-5</sup> in 1000 iterations, with a slight improvement to the precision of the predictions. Notably, the convergence begins at a simiar iteration, but once begun, convergence happens at a faster rate. For any given required accuracy, the result is obtained with fewer iterations.

## Binary output encoding
Now we come to the reason I decided to do this investigation. In my googling, I found that rule 2 from the above list was quietly controversial. The most common suggestion I found on ANN and ML fora, was to encode binary outputs with a single node (1 of C-1 encoding), using 0 for false, 1 for true. With a logistic sigmoid output activation function, results are guaranteed to be between 0 and 1, and can be interpreted as probabilities: an outcome of 0.3 indicates a 30% probability that the assertion is true. However, LeCun explicitly advises against this encoding. 
> First, instabilities can result. The training process will try to drive the output as close as possible to the target values, which can only be achieved asymptotically. As a result, the weights (output and even hidden) are driven to larger and larger values where the sigmoid derivative is close to zero. The very large weights increase the gradients, however, these gradients are then mutiplied by and exponentially small sigmoid derivative ... producing a weight update close to zero. As a result, the weights may become stuck.

> Second, when the outputs saturate, the network gives no indication of confidence level. When an input pattern falls near a decision boundary the output class is uncertain. Ideally this should be reflected in the network by an output value that is in between the possible target values, i.e. not near either asymptote. Howerver, large weights tend to force all outputs to the tails of the sigmoid regardless of the uncertainty. Thus the network may predict a wrong class without giving any indication of its low confidence in the result.

From my perspective it seems likely to be more efficient for the output node to be able to approach the correct answer from both sides than to approach it asymptotically, so LeCun's suggestion makes intuitive sense to me. Instead of the usual approach, LeCun suggests either adjusting the output activation function or rescaling the binary values so that True and False end up at the point of maximum second derivative of the log sigmoid or tanh activation function. This gives ample room toward the asymptote for the solution to overshoot and then backtrack, and also keeps the target values outside of the linear range of the sigmoid (non-linearity is the secret sauce of neural networks).

Having read LeCun's advice, I was surprised to find very little mention of it on the internet. Warren Sarle [explicitly disagrees](ftp://ftp.sas.com/pub/neural/FAQ2.txt) with Lecun:
> The common practice of using target values of .1 and .9 instead of 0 and 1 prevents the outputs of the network from being directly interpretable as posterior probabilities, although it is easy to rescale the outputs to
produce probabilities (Hampshire and Pearlmutter, 1991, figure 3). This practice has also been advocated on the grounds that infinite weights are required to obtain outputs of 0 or 1 from a logistic function, but in fact,
weights of about 10 to 30 will produce outputs close enough to 0 and 1 for all practical purposes, assuming standardized inputs. Large weights will not cause overflow if the activation functions are coded properly;

later:
> The formula for the logistic activation function is often written as: 
  
>  netoutput = 1 / (1+exp(-netinput));

> But this formula can produce floating-point overflow in the exponential function if you program it in this simple form. To avoid overflow, you can do this: 

>  if (netinput < -45) netoutput = 0;

>  else if (netinput > 45) netoutput = 1;

>  else netoutput = 1 / (1+exp(-netinput));

> The constant 45 will work for double precision on all machines that I know of, but there may be some bizarre machines where it will require some adjustment. Other activation functions can be handled similarly.

This argument avoids addressing the preferability of approaching the target from either side, and the suggested fix for the logistic activation function actually makes the slope identically zero for large inputs. The latter point seems to be a recipe for stuck weights. 

When experts disagree, ask random strangers. Searching on Stack Exchange ([e.g.](http://stats.stackexchange.com/questions/196670/targets-of-0-1-0-9-instead-of-0-1-in-neural-networks-and-other-classification-al)) reveals that others have asked the question, but gotten no real answers. Many academic publications ([e.g.](https://www.researchgate.net/publication/3950351_Who_is_afraid_of_the_big_bad_ANN)) and tutorials ([e.g.](http://web.engr.oregonstate.edu/~xfern/classes/cs534/notes/neuralnet-9-11.pdf)) recommend rescaling binary outputs to 0.1 and 0.9 (different values than what LeCun recommends), for similar reasons that LeCun specifies.

Thus I decided to explore the subject on my own. We certainly won't get a definitive answer by analyzing such a simple toy problem, but maybe we'll get some insight.

First, what value of the logistic sigmoid has the strongest curvature?

```python
x = np.arange(-3, 3, 0.00001)
f = 1/(1+np.exp(-x))
fp = np.exp(x)/(1+np.exp(x))**2
fpp = -np.exp(x)*(np.exp(x)-1)/(np.exp(x)+1)**3

max_ind = np.argmax(fpp)
min_ind = np.argmin(fpp)
xmax = x[max_ind]
xmin = x[min_ind]

plt.plot(x, f, label='f(x)')
plt.plot(x, fp, label='f\'(x)')
plt.plot(x, fpp, label='f\"(x)')
plt.legend(loc=2)
plt.title('logistic sigmoid')
plt.xlabel('x')
plt.axvline(x=xmax, ymin=0, ymax=1)
plt.axvline(x=xmin, ymin=0, ymax=1)

print "Maximum curvature at logsig(%06.4f) = %06.4f and \
logsig(%06.4f) = %06.4f"  % (xmax,f[max_ind], xmin,f[min_ind])
```

The vertical lines show x-values of the maximum curvature magnitude. They occur at about x = +/- 1.317, and correspond to sigmoid values conveniently close to 0.2 and 0.8, which we can use for False and True, respectively. So let's train our network with these values. 

We'll try it once with traditional binary inputs:

```python
train_err, val_err = binary_ANN(1,0,0.8,0.2)
```

We now get convergence to an MSE of ~2 x 10<sup>-14</sup> after 1000 iterations. To achieve the same MSE as the first test requires just under 700 iterations.

```python
    plt.semilogy(range(len(train_err)), train_err, label='train')
    plt.semilogy(range(len(val_err)), val_err, label='val')
    plt.title('error')
    plt.legend(loc=3)
    plt.xlabel('epoch')
    plt.ylabel('Mean squared error')
    plt.ylim(1e-4, 1)
    plt.xlim(0, 700)
    print
```

Like the first two tests, this one doesn't start to converge until about 250 iterations, but with this output encoding, once the MSE starts to drop, it freefalls.

Now let's combine our new input encoding with our new output encoding.

```python
train_err, val_err = binary_ANN(1,-1,0.8,0.2)
```

This encoding gives us a similar final MSE, but fully converges in just over 500 iterations. To achieve the same MSE as the first test takes less than 500 iterations.

```python
    plt.semilogy(range(len(train_err)), train_err, label='train')
    plt.semilogy(range(len(val_err)), val_err, label='val')
    plt.title('error')
    plt.legend(loc=3)
    plt.xlabel('epoch')
    plt.ylabel('Mean squared error')
    plt.ylim(1e-4, 1)
    plt.xlim(0,500)
    print
```

## What's the take-home?

For this problem, Zero-centered binary input encoding has a significant impact on the speed of convergence, and non-asymptotic binary output encoding has an enormous effect on the final mean squared error. Whether these effects generalize to more interesting problems has not been demonstrated here. If so, it's clear that the modified input encoding is useful (shorter runtimes), but it's not clear that the improved final convergence is especially helpful for binary classification problems. If we are willing to accept two decimal places of accuracy in a binary output, we can get that level of convergence in the same number of iterations with the either output encoding. If we are interested in probabilities, then perhaps the improved precision can be useful, as long as we properly rescale the results. The original output encoding allows us to interpret the results as probabilities with no additional work. Note also that binary cross-entropy gives better results than MSE as a loss function, and is incompatible with the output rescaling described here, though perhaps it would work ok if we rescaled the output activation function instead of the binary values themselves. I haven't tried that, so I can't say for sure.

## Appendix: Let's try tanh

Perhaps we can do better with a tanh output activation function? This requires only two small adjustments to the script: a tanh output layer, a zero-centered output encoding:


```python
def binary_ANN_tanh(in_true, in_false, out_true, out_false):

    # generate training data 
    #(XOR, or any other binary logic truth table)
    dataModel = [[(in_false, in_false), (out_false)],
                 [(in_false, in_true), (out_true)],
                 [(in_true, in_false), (out_true)],
                 [(in_true, in_true), (out_false)]]


    train_size = 1000     # number of training inputs
    np.random.seed(1)

    # add rows randomly from the truth table
    def make_data(size):
        inp = []
        targ = []
        for ind in range(size):
            inpt, target = dataModel[np.random.randint(4)]
            inp.append(inpt)
            targ.append(target)
        return np.array(inp), np.array(targ).reshape(-1, 1)

    train_in, train_target = make_data(train_size)

    # 1. Define Network:
    num_hidden_nodes = train_in.shape[1] + 1
    model = Sequential()
    # 3 nodes in the hidden layer with tanh activation function
    model.add(Dense(num_hidden_nodes, input_dim=train_in.shape[1], activation='tanh', init='uniform'))
    # 1 output node with tanh activation function
    model.add(Dense(1, activation='tanh', init='uniform'))

    # 2. Compile Network
    model.compile(loss='mse', optimizer='adadelta', metrics=['accuracy'])
    t = timer()

    # 3. Fit Network
    history = model.fit(train_in,
                        train_target,
                        nb_epoch=1000,
                        validation_split=0.2,
                        batch_size=32,
                        verbose=0)

    # How long did training take?
    print 'Time elapsed:', timer() - t

    # Visualize convergence
    %matplotlib inline
    train_err = history.history['loss']
    train_acc = history.history['acc']
    val_err = history.history['val_loss']
    val_acc = history.history['val_acc']

    plt.semilogy(range(len(train_err)), train_err, label='train')
    plt.semilogy(range(len(val_err)), val_err, label='val')
    plt.title('error')
    plt.legend(loc=3)
    plt.xlabel('epoch')
    plt.ylabel('Mean squared error')

    # Make predictions
    demo_in = np.array([element[0] for element in dataModel])
    demo_target = np.array([element[1] for element in dataModel])
    demo_out = model.predict(demo_in)
    misfit = demo_out-demo_target.reshape(-1,1)

    print 'Prediction:'
    for i, val in enumerate(demo_in):
        print demo_in[i], demo_out[i]

    print "\nMisfit:\n", misfit

    print "\nFinal mean squared error:", train_err[-1]
    
    return train_err, val_err

train_err, val_err = binary_ANN_tanh(1,-1,1,-1)
```

The strongest curvature for tanh occurs at tanh(-0.6580) = -0.5774 and tanh(0.6580) = 0.5774.

```python
x = np.arange(-3, 3, 0.00001)
f = np.tanh(x)
fp = 1/np.cosh(x)**2
fpp = -2*np.tanh(x)/np.cosh(x)**2

max_ind = np.argmax(fpp)
min_ind = np.argmin(fpp)
xmax = x[max_ind]
xmin = x[min_ind]

plt.plot(x, f, label='f(x)')
plt.plot(x, fp, label='f\'(x)')
plt.plot(x, fpp, label='f\"(x)')
plt.legend(loc=2)
plt.title('tanh')
plt.xlabel('x')
plt.axvline(x=xmax, ymin=0, ymax=1)
plt.axvline(x=xmin, ymin=0, ymax=1)

print "Maximum curvature at tanh(%06.4f) = %06.4f and \
tanh(%06.4f) = %06.4f"  % (xmax,f[max_ind], xmin,f[min_ind]) 
```

```python
train_err, val_err = binary_ANN_tanh(1,-1,0.5774,-0.5774)
```

The results are reasonably similar to those for logistic sigmoid output activation.

## Disclaimer
Note that the tests described here only apply to neural nets with binary outputs. Classification problems with multiple output classes traditionally use one-hot (aka 1 of C) encoding for their output values and employ a softmax activation function, which sums all of the output node values to 1. Rescaled targets would not be appropriate for such formulations. An alternative approach in that case would be to rescale the softmax function itself. I'll leave that to someone else for now.
